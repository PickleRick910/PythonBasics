```python
from itertools import filterfalse
## nested tuple.
n_t=((1,2,3,4),("a","b","c"),(True,False))

## access the element in tuple
print(n_t[0])
```

```python
for sub_tuple in n_t:
  for item in sub_tuple:
    print(item,end="")
  print()
```
