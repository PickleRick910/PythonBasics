```python
#while
##continues to execute as log as condition is true
count=0
while count%2==0:
  print(count)
  count=count+1
```

```python
##beak
#breaks loops prematurely
for i in range(10):
  if i ==5:
    break
  print(i)
```

```python
##continue
##skips current iteration and continue with next
for i in range(10):
  if i%2 ==0:
    continue
  print(i)
```

```python
###pass
##doesnt do anything
for i in range(6):
  if i ==3:
    print("the number is",i)
    pass
  print(i)
```

```python
##nested loops
for i in range(3):
  for j in range(5):
    print(f"i:{i} and j:{j}")
```
