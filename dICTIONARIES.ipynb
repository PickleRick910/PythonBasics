```python
##nested dict
##dict inside other dict
students={
    "stu1":{"name":"krish","age":32},
    "stu2":{"name":"peter", "age":35}
}
print(students)
```

```python
##accessing nested dic
print(students["stu2"]["name"])
print(students["stu2"]["age"])
```

```python
###iterating over nested cdicts
for student_id,student_info in students.items():
  print(f"{student_id}:{student_info}")
  for key,value in student_info.items():
    print(f"{key}:{value}")
```

```python
###dict comprehnetion
sq={ x:x**2 for x in range(5)}
print(sq)
```

```python
##conditonal dict comprehention
even={x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
##conditonal dict comprehention
odd={x:x**2 for x in range(10) if x%3==0}
print(odd)
```

```python
##practical ex
##use dict to count the freq of elemtns in list

num=[1,2,2,2,2,2,3,4,5,2,3,4,5]
freq={}
for number in num:
  if number in freq:
    freq[number]+=1
  else:
    freq[number]=1
print(freq)
```
