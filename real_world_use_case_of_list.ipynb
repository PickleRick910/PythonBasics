```python
###manage a to do list
to_do_list=["buy groceries", "clean the house", "pay bills"]
## adding to task
to_do_list.append("schedule a meeting")
to_do_list.append("go for run")
### remove a task
to_do_list.remove("clean the house")

##checking if a task in list
if"pay bills" in to_do_list:
  print("dont forget to pay bill utility")

print("to do list remaining")
for task in to_do_list:
  print(f"-{task}")
```

```python
####
grades=[85,92,90,88]

grades.append(98)
##calc avg
avg_grade=sum(grades)/len(grades)
print((f"avg grade: {avg_grade:.2f}"))
##finding the highest and lloweest grade
h_grade=max(grades)
l_grade=min(grades)
print(l_grade)
print(h_grade)
```
